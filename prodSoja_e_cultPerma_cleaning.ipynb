```python
import pandas as pd
```

```python
url_soja = 'https://github.com/LABSensing/LabSesing-case-study/raw/refs/heads/main/Producao-Soja/produ%C3%A7%C3%A3osoja.xls'
```

```python
sojadf = pd.read_excel(url_soja)
sojadf.info()
```

```python
estados = ['Acre', 'Amapá', 'Amazonas', 'Maranhão', 'Mato Grosso', 'Pará', 'Roraima', 'Rondônia', 'Tocantins']
sojadf = sojadf.query(' @estados in Estado')
sojadf
```

```python
## verificando quantos valors nulos ainda sobraram
sojadf.info()
```

```python
soja_df = sojadf.drop('Codigo', axis = 1)
soja_df.to_csv('producaosojaClean', index = False)
```

plot

```python
import matplotlib.pyplot as plt
```

```python
anos_str = [str(year) for year in range(1999, 2021)]
cores_estado = {}
plt.figure(figsize=(18, 10))

for i, estado in enumerate(soja_df['Estado'].unique()):
    dados_estado = soja_df[soja_df['Estado'] == estado][anos_str].iloc[0]
    linha, = plt.plot(anos_str, dados_estado, marker='o', linestyle='-', label=estado)
    cores_estado[estado] = linha.get_color()  # Salvar a cor atribuída automaticamente pelo matplotlib

plt.xlabel("Year", fontsize=14)
plt.ylabel("Soy production (R$, at 2010 prices (thousands))", fontsize=14)
plt.grid(True, which="both", linestyle="--", linewidth=0.5)

# Ajustar a escala do eixo Y para log base 10
plt.yscale('log')

# Exibir o gráfico
plt.show()
```

✅ Área de Culturas Permanentes (hectares)

```python
url_harea = 'https://github.com/LABSensing/LabSesing-case-study/raw/refs/heads/main/culturasPermanente/harvestedarea.xls'
```

```python
hadf = pd.read_excel(url_harea)
hadf.info()
```

```python
hadf.head()
```

```python
hadf_estados = hadf.query(' @estados in Estado')
hadf_estados
```

```python
##hadf_estados.drop('Codigo', axis = 1, inplace = True)
hadf_estados
```

```python
hadf_estados.to_csv('colheitapermanenteClean', index = False)
```

plot

pd.melt() = transforma o formato de uma tabela,


```python
#converte a tabela de formato wide para long, tornando a area colhida em uma coluna
hadf_long = hadf_estados.melt(id_vars=["Estado", "Sigla", "Codigo"], var_name="Ano", value_name="Area_Colhida")
hadf_long
```

```python
hadf_long["Area_Colhida"] = pd.to_numeric(hadf_long["Area_Colhida"], errors="coerce") / 1000  # Converter para milhares de hectares
```

```python
estados = hadf_long["Estado"].unique()
num_estados = len(estados)
cols = 3  # Número de colunas desejado
rows = -(-num_estados // cols)  # Cálculo para determinar o número de linhas
fig, axes = plt.subplots(rows, cols, figsize=(cols * 6, rows * 3))
axes = axes.flatten()

for i, estado in enumerate(estados):
    subset = hadf_long[hadf_long["Estado"] == estado]

    cor = cores_estado.get(estado, "gray")  # Usa a cor salva ou cinza como fallback

    axes[i].fill_between(subset["Ano"], subset["Area_Colhida"], alpha=0.7, color=cor)
    axes[i].set_xlabel("Ano")
    axes[i].set_ylabel("Área Colhida (milhares de ha)")
    axes[i].set_title(f"Evolução da Área Colhida - {estado} (1999-2020)")
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```
